1. Batch Learning (Offline Learning)
The model is trained using the entire dataset at once or in large batches.
Once trained, the model does not update until a new batch of data is collected and retrained.
It is computationally expensive and requires significant resources.
Suitable for applications where data changes slowly, like fraud detection, medical diagnosis, or recommendation systems.
Example: Training a deep learning model on historical data for image classification.
2. Online Learning (Incremental Learning)
The model is updated continuously as new data arrives, without needing retraining on the entire dataset.
Works well in dynamic environments where data distribution changes over time (concept drift).
Can handle real-time data streams efficiently.
Suitable for applications like stock market predictions, spam detection, and real-time recommendation systems.
Example: Updating a recommendation engine based on recent user interactions.

```python
import numpy as np
from sklearn import linear_model

n_samples, n_features = 1, 500

y = np.random.randn(n_samples)
X = np.random.randn(n_samples, n_features)

clf = linear_model.SGDRegressor()

import time

start_time = time.time()

clf.partial_fit(X, y)

elapsed_time = time.time() - start_time

print(elapsed_time)
```
